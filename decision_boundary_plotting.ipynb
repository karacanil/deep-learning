```python
#Decision Boundary Plotting Demo

#Import relevant libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Declare an example function
def f(x, y):
    return 2*np.square(x) + 5*np.square(y)

#Define a more complicated example function
def g(x, y):
    return -(y*np.square(x)+(1-y)*np.square(1-x))
```

```python
#Define domain of interest(doi)
x_doi = (-5, 5)
y_doi = (-5, 5)
grid_size = 0.1
```

```python
#Generate the mesh points
x_mesh, y_mesh = np.meshgrid(np.arange(x_doi[0], x_doi[1], grid_size),
                            np.arange(y_doi[0], y_doi[1], grid_size))
```

```python
#Calculate the value of f(x,y)
#z = f(x_mesh.ravel(), y_mesh.ravel())
z = g(x_mesh.ravel(), y_mesh.ravel())
```

```python
#Reshape the f(x,y)
z = z.reshape(x_mesh.shape)
```

```python
#Plot the contour
plt.contour(x_mesh, y_mesh, z, levels=[1, 5, 9, 13, 17, 29], cmap=plt.cm.Spectral)
```
